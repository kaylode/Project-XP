```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.utils.data as data
from torch.autograd import Variable
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from torchvision import datasets, transforms, models
from torchsummary import summary
```

```python
device = torch.device("cuda: 0" if torch.cuda.is_available() else "cpu")
```

```python
IMG_SIZE = 32
```

```python
DATA_DIR = {
    "train": "data"
}
```

```python
train_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(IMG_SIZE, padding = 2),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

val_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
```

```python
trainset = datasets.CIFAR10(DATA_DIR["train"],
                            train= True,
                            transform=train_transform,
                            download = True)
testset = datasets.CIFAR10(DATA_DIR["train"],
                            train= False,
                            transform=val_transform,
                            download = True)
```

```python
CLASSES = trainset.classes
NUM_CLASSES = len(CLASSES)
```

```python
print(f"Number of training samples: {len(trainset)}")
print(f"Number of vaidating samples: {len(testset)}")
print(f"Number of unique labels : {NUM_CLASSES}")
print(f"Unique labels name: \n{CLASSES}")
```

```python
# Visualize some training samples
from mpl_toolkits.axes_grid1 import ImageGrid

def visualize_samples(NUM_PIC=10):
    a = np.random.randint(100-NUM_PIC)
    fig = plt.figure(figsize=(NUM_PIC*2,NUM_PIC*2))
    plt.title("Training Samples", fontsize=25)
    plt.xticks([])
    plt.yticks(np.arange(10)[-1:-11:-1]+0.3, CLASSES, fontsize=25)
    grid = ImageGrid(fig, 111, (NUM_PIC,NUM_PIC), axes_pad=0)
    
    imgs_id_list = []
    for i in range(10):
        imgs_id = np.where(np.array(trainset.targets) == i)
        imgs_id_list += list(imgs_id[0][a:a+NUM_PIC])
    
    
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    
    for ax, im_id in zip(grid, imgs_id_list):
        img = trainset[im_id][0]
        img = img.numpy().transpose((1,2,0))
        img = (img*std + mean)
        img = np.clip(img,0,1)
        ax.imshow(img)
        ax.axis("off")
    
    
    plt.show()
    
    
visualize_samples()
```

```python
def visualize_frequency():
    id_count = {CLASSES[i]: len(np.where(np.array(trainset.targets) == i)[0]) for i in range(10)}

    fig = plt.subplots(figsize=(15,10))
    plt.bar(id_count.keys(), height=id_count.values())
    plt.title("Frequency of each class", fontsize= 20)
    plt.xlabel("Class", fontsize= 20)
    plt.ylabel("Number of samples", fontsize= 20)
    plt.show()
    
visualize_frequency()    
```

```python
# Load pretrained model

model = models.resnet50(pretrained=True)
for params in model.parameters():
    params.requires_grad = False
in_features = model.fc.in_features
model.fc = nn.Linear(in_features,10)

model.load_state_dict(torch.load("model/resnet50-0.5.pt"))
modules=list(model.children())[:-1]
model=nn.Sequential(*modules)
model = model.to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=2e-4, weight_decay = 1e-5, betas=(0.5, 0.999))
criterion = nn.CrossEntropyLoss()

summary(model, (3, 32, 32))
```

```python
BATCH_SIZE = 128
workers = 2
trainloader = data.DataLoader(trainset,
                              shuffle = True,
                              batch_size=BATCH_SIZE,
                              num_workers=workers)
testloader = data.DataLoader(testset,
                              batch_size=BATCH_SIZE,
                              num_workers=workers)
```

```python
def train_step(model, optimizer, criterion, trainloader):
    model.train()
    total_loss = 0
    correct = 0
    total = 0
    
    for i, (img,label) in enumerate(trainloader):
        optimizer.zero_grad()
        
        img = Variable(img).view(-1,3,IMG_SIZE,IMG_SIZE).to(device)
        label = Variable(label).to(device)
        
        output = model(img)
        loss = criterion(output,label)
        total_loss += loss.data
        loss.backward()
        optimizer.step()
        
        predictions = torch.max(output, 1)[1]
        correct += (predictions==label).sum()
        total += len(label)
        
    accuracy = correct*1.0/total
    return total_loss, accuracy
```

```python
def evaluate_step(model, criterion, testloader):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0
    
    with torch.no_grad():
        for i, (img,label) in enumerate(testloader):
            
            img = Variable(img).view(-1,3,IMG_SIZE,IMG_SIZE).to(device)
            label = Variable(label).to(device)
            
            output = model(img)
            loss = criterion(output,label)

            total_loss += loss.data
            predictions = torch.max(output, 1)[1]
            correct += (predictions==label).sum()
            total += len(label)
        accuracy = correct*1.0/total
    return total_loss, accuracy
```

```python
TRAIN_DATA = 1

loss_list = {
    "train": [],
    "val": []
}

acc_list = {
    "train": [],
    "val": []
}


if TRAIN_DATA:
    EPOCHS = 100
    print("Start training...")
    for epoch in range(EPOCHS):
        train_loss, train_acc = train_step(model, optimizer, criterion,trainloader)
        val_loss, val_acc = evaluate_step(model, criterion,testloader)
        loss_list["train"].append(train_loss)
        loss_list["val"].append(val_loss)
        acc_list["train"].append(train_acc)
        acc_list["val"].append(val_acc)
        print(f"Epoch: [{epoch+1}|{EPOCHS}] \n\t Training Loss: {train_loss:.3f} | Training Acc: {train_acc:.3f}\n\t Validating Loss: {val_loss:.3f} | Validating Acc: {val_acc:.3f}")
        torch.save(model.state_dict(), "model/modelVGG19.pt")
    print("Training Completed!")
```

```python
def plot(loss_list, acc_list, EPOCHS):
    fig = plt.figure(figsize=(20,5))
    fig.add_subplot(121)
    plt.plot(range(EPOCHS), loss_list["train"], label = "Training Loss")
    plt.plot(range(EPOCHS), loss_list["val"], label = "Validating Loss", color="orange")
    plt.title("Loss per epoch")
    
    fig.add_subplot(122)
    plt.plot(range(EPOCHS), acc_list["train"], label = "Training Accuracy")
    plt.plot(range(EPOCHS), acc_list["val"], label = "Validating Accuracy", color="orange")
    plt.title("Accuracy per epoch")
    plt.show()

plot(loss_list, acc_list, EPOCHS)
```
